```python
# Import Libraries
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
# Read CSV
data = pd.read_csv('./data/german.data', sep=" ", names=['Checking_acc', 'duration_mnth', 'Credit_hist', 
                'Purpose', 'credit_amt', 'Saving_acc', 'Emp_status','installment_rate','Pstatus_and_sex',
                'Guarantor','Residence_LOS', 'Property', 'age', 'Installment_plan', 'Housing',
                 'existing_credit', 'Job', 'dependents', 'Telephone', 'Foreign_worker', 'credit'])
print(data.shape) 
data.to_csv('./data/GermanCredit_num.csv', index=False)
data.head()
```

#### Creating a dictionary for each features

```python
# Checking Account feature
checking_acc_dict = {
    'code1': ['A11', 'A12', 'A13', 'A14'],
    'checking_acc': ['Less than 0 DM', 'Btw 0 to 199 DM', 'Equal or Greater than 200 DM', 'none']
}

checking_acc_dict = pd.DataFrame(checking_acc_dict)
checking_acc_dict
```

```python
# Credit History Feature
credit_hist_dict = {
    'code2': ['A30','A31','A32','A33','A34'],
    'credit_hist': ['No credits taken','All credits taken paid back duly', 'Existing credits paid back duly',
                'Delays in paying off','Critical account']
}
credit_hist_dict = pd.DataFrame(credit_hist_dict)
credit_hist_dict   
```

```python
# Purpose  Feature
purpose_dict = {
    'code3': ['A40','A41', 'A42', 'A43', 'A44', 'A45', 'A46', 'A47', 'A48', 'A49', 'A410' ],
    'purpose': ['new car', 'used car', 'furniture/equipment', 'radio/television', 'domestic appliances',
                'repairs', 'education', 'vacation', 'retraining', 'business', 'others']
}
purpose_dict= pd.DataFrame(purpose_dict)
purpose_dict
```

```python
# Savings account Feature
savings_acc_dict = {
    'code4': ['A61', 'A62', 'A63', 'A64', 'A65'],
    'savings_acc': ['Less than 100 DM','Btw 100 to 499 DM','Btw 500 to 999 DM','Equal or Greater than 1000 DM', 'none']
}
savings_acc_dict= pd.DataFrame(savings_acc_dict)
savings_acc_dict
```

```python
# Employment Status  Feature
emp_status_dict = {
    'code5': ['A71', 'A72', 'A73', 'A74', 'A75'],
    'emp_status': ['unemployed', 'Less than a year', 'Btw 1 to 4 years', 'Btw 4 to 7 years', 'Greater than 7 years']
}
emp_status_dict= pd.DataFrame(emp_status_dict)
emp_status_dict
```

```python
# Personal Status and sex  Feature
pstatus_sex_dict = {
    'code6': ['A91', 'A92', 'A93', 'A94', 'A95'],
    'pstatus_sex': [ 'male_divorced/separated', 'female_divorced/separated/married','male_single', 'male_married/widowed',
                'female_single']   
}
pstatus_sex_dict= pd.DataFrame(pstatus_sex_dict)
pstatus_sex_dict
```

```python
# Guarantor Feature
guarantor_dict = {
    'code7': ['A101', 'A102', 'A103'],
    'guarantor': ['none','co-applicant','guarantor']   
}
guarantor_dict= pd.DataFrame(guarantor_dict)
guarantor_dict
```

```python
# Property Feature
property_dict = {
    'code8': ['A121', 'A122', 'A123', 'A124'],
    'property': ['real estate','BSSA or life insurance', 'car or other', 'none']       
}
property_dict= pd.DataFrame(property_dict)
property_dict
```

```python
# Installment Plan Feature
instl_plan_dict = {
    'code9': ['A141', 'A142', 'A143'],
    'installment_plan': ['bank','stores', 'none']  
}
instl_plan_dict= pd.DataFrame(instl_plan_dict)
instl_plan_dict
```

```python
# Housing Feature
housing_dict = {
    'code10': ['A151', 'A152', 'A153'],
    'housing': ['rent','own', 'free']  
}
housing_dict= pd.DataFrame(housing_dict)
housing_dict
```

```python
# Job Feature
job_dict = {
    'code11': ['A171', 'A172', 'A173', 'A174'],
    'job': ['unemployed/unskilled - non-resident','unskilled - resident','skilled employee',
                'self-employed/management employee']
}
job_dict= pd.DataFrame(job_dict)
job_dict
```

```python
# telephone
telephone_dict = {
    'code12':['A191', 'A192'],
    'telephone': ['none','registered']
}
telephone_dict= pd.DataFrame(telephone_dict)
telephone_dict
```

```python
# Foreign worker Feature
foreign_worker_dict = {
    'code13':['A201', 'A202'],
    'foreign_worker': ['yes','no']
}
foreign_worker_dict= pd.DataFrame(foreign_worker_dict)
foreign_worker_dict
```

#### Merging Data and Dropping columns

```python
# Merging all the columns
merged_df = pd.merge(data, checking_acc_dict, left_on='Checking_acc', right_on='code1', how='left')
merged_df = pd.merge(merged_df, credit_hist_dict, left_on='Credit_hist', right_on='code2', how='left')
merged_df = pd.merge(merged_df, purpose_dict, left_on='Purpose', right_on='code3', how='left')
merged_df = pd.merge(merged_df, savings_acc_dict, left_on='Saving_acc', right_on='code4', how='left')
merged_df = pd.merge(merged_df, emp_status_dict, left_on='Emp_status', right_on='code5', how='left')
merged_df = pd.merge(merged_df, pstatus_sex_dict, left_on='Pstatus_and_sex', right_on='code6', how='left')
merged_df = pd.merge(merged_df, guarantor_dict, left_on='Guarantor', right_on='code7', how='left')
merged_df = pd.merge(merged_df, property_dict, left_on='Property', right_on='code8', how='left')
merged_df = pd.merge(merged_df, instl_plan_dict, left_on='Installment_plan', right_on='code9', how='left')
merged_df = pd.merge(merged_df, housing_dict, left_on='Housing', right_on='code10', how='left')
merged_df = pd.merge(merged_df, job_dict, left_on='Job', right_on='code11', how='left')
merged_df = pd.merge(merged_df, telephone_dict, left_on='Telephone', right_on='code12', how='left')
merged_df = pd.merge(merged_df, foreign_worker_dict, left_on='Foreign_worker', right_on='code13', how='left')
merged_df.head(5)
```

```python
# Dropping repeated data
merged_df.drop(['code1','code2','code3', 'code4', 'code5', 'code6','code7', 'code8', 'code9', 'code10',
               'code11', 'code12', 'code13', 'Checking_acc', 'Credit_hist', 'Purpose', 'Saving_acc',
               'Emp_status', 'Pstatus_and_sex', 'Guarantor','Property', 'Installment_plan', 'Housing',
               'Job', 'Telephone', 'Foreign_worker'], axis=1, inplace=True)
```

```python
# Arranging the columns to suit the original data
df = merged_df[['checking_acc', 'duration_mnth', 'credit_hist', 'purpose', 'credit_amt',
           'savings_acc', 'emp_status',  'installment_rate', 'pstatus_sex', 'guarantor',
           'Residence_LOS', 'property', 'age', 'installment_plan', 'housing', 'existing_credit',
           'job', 'dependents', 'telephone', 'foreign_worker','credit']]
```

```python
df.head()
```

```python
# Fixing the target columns
df['credit'] = df['credit'].replace({1: 'bad', 2: 'good'})
```
